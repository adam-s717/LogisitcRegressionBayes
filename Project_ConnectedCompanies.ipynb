This file creates a weighted adjencey matrix between all companies' returns in the S&P500 (only the ones with no missing data were considered) using the Spearman Rank correlation. We then convert this to an adjencey matrix and finally identify the 5 most connect companies. 

```python
AllReturns = pd.read_csv('sp500_joined_returns.csv')
```

```python
##Obtain list of company names present in the pandas dataframe
compNames = list(AllReturns)

##Create an empty list for companies we will use
useComps = []

##Now determine which stocks do not have any missing values in the columns
for i in compNames:
    
    ##Boolean to asses if there are any missing values in the column 
    if AllReturns[i].isnull().values.any() == False:
        
        ##If there are not missing values append to a new list
        useComps.append(i)
```

```python
useComps.pop(0)
```

```python
stocks = AllReturns[useComps]
```

```python
stocks.head()
```

```python
WA = stocks.corr(method='spearman').fillna(0)
```

```python
WA
```

```python
WA.where(WA <= .6, 1,inplace=True) #If its greater than .6 replace with 1
WA.where(WA == 1, 0,inplace=True)  #If it doest equal 1 replace with 0
Adjencey = WA
```

```python
Adjencey
```

```python
Sums = Adjencey.sum().loc[lambda x: x>1]
```
